## 7. Konsol Menusu (Interaktif Kullanim)

Asagidaki hucre calistirildiginda tam konsol menusu baslar.

```python
def uygulamayi_calistir():
    """Ana konsol dongusu. Menu uzerinden tum islemlere erisim saglanir."""
    global gelirler, giderler
    gelirler, giderler = csv_oku(CSV_DOSYA)

    while True:
        menu_goster()
        secim = input().strip()

        if secim == '1':
            gelir_ekle(gelirler)
        elif secim == '2':
            gider_ekle(giderler)
        elif secim == '3':
            islemleri_listele(gelirler, giderler)
        elif secim == '4':
            df = verileri_dataframe_yap(gelirler, giderler)
            if not df.empty:
                toplam_gelir_gider(df)
                aylik_analiz(df)
                numpy_istatistik(df)
        elif secim == '5':
            df = verileri_dataframe_yap(gelirler, giderler)
            if not df.empty:
                aylik_grafik(df)
                gelir_gider_bar(df)
                pasta_grafik(df)
        elif secim == '6':
            csv_kaydet(CSV_DOSYA, gelirler, giderler)
        elif secim == '7':
            csv_kaydet(CSV_DOSYA, gelirler, giderler)
            print('Gule gule! Veriler kaydedildi.')
            break
        else:
            print('Gecersiz secim. Lutfen 1-7 arasinda bir deger girin.')

# Menuyu baslatmak icin su satiri yorumdan cikarin:
# uygulamayi_calistir()
print('Menuyu baslatmak icin: uygulamayi_calistir()')
```
